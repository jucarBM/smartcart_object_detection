Extrayendo el zip

```python
import os
import zipfile
local_zip = "dataset_filtrado.zip"
zip_ref = zipfile.ZipFile(local_zip,'r')
zip_ref.extractall("dataset")
zip_ref.close()
```

Genreando el dataset de manera randomica

```python
path = "dataset/dataset_filtrado"
content = os.listdir(path)
```

```python
path_general = "datasetFinal"
path_train = f'{path_general}/train' 
path_test = f'{path_general}/test'

os.mkdir(path_general)
os.mkdir(path_train)
os.mkdir(path_test)
```

```python
import random
import shutil

train = 0.7
for nCount in range(int(len(content)*train)):
  random_choice_img = random.choice(content)
  random_choice_img_abs = f'{path}/{random_choice_img}'
  target_img = f'{path_train}/{random_choice_img}'
  shutil.copyfile(random_choice_img_abs, target_img)
  content.remove(random_choice_img)
```

```python
for img in content:
  random_choice_img_abs = f'{path}/{img}'
  target_img = f'{path_test}/{img}'
  shutil.copyfile(random_choice_img_abs, target_img)
```
